##4. Data Cleaning

```python
data.isnull().sum()

#droping all missing values rows
data.dropna(inplace=True)

#data.fillna()
data.isnull().sum()
```

```python
data.duplicated()
```

```python
# Dealing with duplicates rows
data.duplicated().sum()

#Deleting all duplicated row
#data[data.duplicated()]

data.drop_duplicates(inplace=True)
#data.duplicated().sum()

data[data.duplicated()]
```

```python
#Type casting
#changing data types
data.dtypes

data['id']=data['id'].astype(object)
data.dtypes

data['host_id']=data['host_id'].astype(object)
data.dtypes
```

## 5.Data Analysis

Univariat Analysis

```python
# identify outliers in price
df=data[data['price'] < 1500]
sns.boxplot(data=data,x='price')
```

```python
 # Price distribution

plt.figure(figsize=(10,5))
sns.histplot(data=df, x='price')
```

```python
#Price Distribution
plt.figure(figsize=(6,3))
sns.histplot(data=df,x="availability_365")
plt.title("availability_365 Distribution")
plt.ylabel("Frequency")
plt.show()
```

```python
data.dtypes
```

```python
#Mean price by neighborhood.
df.groupby(by='neighbourhood_group')["price"].mean()
```

## Feature Engineering

```python
# price per bad
df['price per bed']=df['price']/df['beds']
df.head()
```

```python
#Average Price per bed
df.groupby(by='neighbourhood_group')["price per bed"].mean()
```

```python
df.head()
```

## Bi Variable Analysis

```python
df.columns
```

```python
# price dependency on neighbourhood
sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type')
```

```python
# number of reviews and price
plt.figure(figsize=(10,5))
plt.title('locality and Review Dependency')
sns.scatterplot(data= df, x='number_of_reviews',y='price',hue='room_type')
```

```python
df.dtypes
```

```python
#Pairplot of price, nights, reviews, availability, by room type.
sns.pairplot(data=df, vars=['price','minimum_nights','number_of_reviews','availability_365'], hue='room_type')
```

```python
#Geographical Distribution of AirBnb Listing
plt.figure(figsize=(10,7))
plt.title('Geographical Distribution of AirBnb Listing')
sns.scatterplot(data=df, x='longitude',y='latitude', hue='room_type')
```

```python
df.dtypes
```

```python
#heat map- correlation of one varibel with other for numerical colum
corr=df[['latitude','longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month','availability_365','beds']].corr()
corr
plt.figure(figsize=(10,5))
sns.heatmap(data=corr,annot=True)
```
